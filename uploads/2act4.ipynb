```python
from tkinter import *
import tkinter.messagebox as tm
from os import path
import csv

#init frame
win = Tk()
win.title('Problem 1')

#methods
def equivalence(tg):
    if tg >= 97 and tg < 100:
        return 1.00
    elif tg >= 94:
        return 1.25
    elif tg >= 91:
        return 1.5
    elif tg >= 88:
        return 1.75
    elif tg >= 85:
        return 2.00
    elif tg >= 82:
        return 2.25
    elif tg >= 79:
        return 2.5
    elif tg >= 76:
        return 2.75
    elif tg >= 75:
        return 3.00
    elif tg >= 74:
        return 3.25
    elif tg >= 72:
        return 3.5
    elif tg >= 70:
        return 3.75
    elif tg >= 65:
        return 4.00
    elif tg >= 60:
        return 4.25
    elif tg >= 58:
        return 4.5
    elif tg >= 50:
        return 5.00

def onSave():
    #file appending and writing
    try:
        if not path.exists('activity3.csv'):
            fp = open('activity3.csv', 'w', newline='')
            fprint = csv.writer(fp)
            fprint.writerow(['Student name', 'Score', 'Items'])

        #appending to file
        fp = open('activity3.csv', 'a', newline='')
        fprint = csv.writer(fp)

        #extracting inputs
        sName = winList[0].get() + ', ' + winList[1].get()
        score = int(winList[2].get())
        items = int(winList[3].get())
        
        if score > items: raise Exception('Score is more than the items')
        if winList[0].get() == '' or winList[1].get() == '': raise Exception('Input all names')
        
        #printing comptutations
        tg = score * 50 / items + 50
        winList[4].set(f'{tg:.2f}')
        
        eq = equivalence(tg)
        winList[5].set(eq)
        
        fprint.writerow([sName,  score, items])
        
        fp.close()     
    except Exception as z:
        tm.showerror('Error:', z)
    

def onClear():
    for i in winList:
        i.set('')

def onView():
    #file reading
    fp = open('activity3.csv', 'r')
    fscan = csv.reader(fp)
    next(fscan)
    
    #prints content
    print('# \tStudent name \tScore \tItems \tTransmuted Grade \tEquivalent')
    count = 1
    for i in fscan:
        tg = int(i[1]) * 50 / int(i[2]) + 50
        eq = equivalence(tg)
        print(count, '\t' + i[0] +'\t' + i[1] + '\t' + i[2] + '\t' + f'{tg:.2f}' + '\t\t\t' + f'{eq:.2f}')
        count += 1
    
    fp.close()

#init contents
lblList = ['Last name', 'First name', 'Score', 'Items', 'Transmuted Grade', 'Equivalent']
initSpan = 2
initWidth = 23
initState = 'normal'

#define components
winList = [StringVar(), StringVar(), StringVar(), StringVar(), StringVar(), StringVar(), StringVar()]

#print gui contents
#print label and field
for i in range(6):
    #span condition
    if i > 1:
        initSpan = 1
        initWidth = 10
    if i > 3: 
        initState = 'disabled'
        
    #contents
    Label(win, text = lblList[i]).grid(row = i, column = 0)
    Entry(win, textvariable = winList[i], width = initWidth, state = initState).grid(row = i, column = 1, columnspan = initSpan)

Button(win, text = 'Save', width = 10, command = onSave).grid(row = 6, column = 0)
Button(win, text = 'Clear', width = 10, command = onClear).grid(row = 6, column = 1)
Button(win, text = 'View', width = 10, command = onView).grid(row = 6, column = 2)

win.mainloop()
```

```python
from tkinter import *

#init frame
win = Tk()
win.title('Check Box')

chk50 = IntVar()
chk60 = IntVar()
txtTotal = StringVar()

def onChk():
    global total
    p50 = chk50.get()
    p60 = chk60.get()
    t50 = 50 if p50 == 1 else 0
    t60 = 60 if p60 == 1 else 0
    total = t50 + t60

def onTotal():
    txtTotal.set(total)

Checkbutton(win, text = 'VM1 - P50', variable = chk50, command = onChk).grid(row = 0, column = 0)
Checkbutton(win, text = 'VM2 - P60', variable = chk60, command = onChk).grid(row = 0, column = 1)
Entry(win, textvariable = txtTotal).grid(row = 1, column = 0)
Button(win, text = 'Total', command = onTotal).grid(row = 2, column = 0)

win.mainloop()
```

```python
from tkinter import *
import tkinter.messagebox as tm

#init frame
win = Tk()
win.title('Menu')

#define components
winList = [StringVar(), StringVar(), StringVar()]
lblList = ['First number', 'Second number', 'Result']

def onPlus():
    try: winList[2].set(int(winList[0].get()) + int(winList[1].get()))
    except Exception as z: tm.showerror('Error:', z)

def onMinus():
    try: winList[2].set(int(winList[0].get()) - int(winList[1].get()))
    except Exception as z: tm.showerror('Error:', z)

def onMultiply():
    try: winList[2].set(int(winList[0].get()) * int(winList[1].get()))
    except Exception as z: tm.showerror('Error:', z)

def onDivide():
    try: winList[2].set(f'{float(winList[0].get()) / float(winList[1].get()):.2f}')
    except Exception as z: tm.showerror('Error:', z)
def onClear():
    for i in winList:
        i.set('')

#print gui
for i in range(2):
    #contents
    Label(win, text = lblList[i]).grid(row = i, column = 0)
    Entry(win, textvariable = winList[i]).grid(row = i, column = 1, columnspan = 3)

#print gui result
Label(win, text = lblList[2]).grid(row = 2, column = 0)
Entry(win, textvariable = winList[2], state = 'disabled').grid(row = 2, column = 1, columnspan = 3)

#button set 1
Button(win, text = '+', command = onPlus).grid(row = 3, column = 0)
Button(win, text = '-', command = onMinus).grid(row = 3, column = 1)
Button(win, text = 'Clear', command = onClear).grid(row = 3, column = 2)

#button set  2
Button(win, text = '*', command = onMultiply).grid(row = 4, column = 0)
Button(win, text = '/', command = onDivide).grid(row = 4, column = 1)
Button(win, text = 'Exit', command = win.destroy).grid(row = 4, column = 2)


win.mainloop()
```
